```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def cast_convert(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<10:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(cast_convert)
movies.head()
```

```python
# movies.iloc[0].crew
```

```python
def director_convert(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(director_convert)
movies.head()
```

```python
def collapse(lists):
    L=[]
    for i in lists:
        i=i.replace(" ","")
        L.append(i)
    return L
```

```python
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)

movies.head()
```

```python
movies["overview"]=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["keywords"]+movies["cast"]+movies["crew"]
movies.head()
```
